### SQL NOTLARI

**SELECT** : Bir veya daha fazla tablodan kayıt almak için kullanılır.

**FROM** : Sorgu için gereken tabloları ve tüm birleştirme bilgilerini listelemek için kullanılır.FROM yan tümcesinde listelenen en az bir tablo olmalıdır.

**WHERE** : Bir SELECT , INSERT , UPDATE veya DELETE deyiminden gelen sonuçları filtrelemek için kullanılır.



#### Örnek:
Kullanıcılar tablosunda yaşı 18 ve 18 den büyük kullanıcıların adlarını listele.

SELECT ad 
FROM kullanicilar

WHERE yas >= 18;

Bir tablodan tüm alanları seçmek için  SELECT * dememiz gerekir.

**ORDER BY** : Normal akışı değiştirmek için kullanılır. ORDER BY yan tümcesi yalnızca SELECT deyimlerinde kullanılabilir. ORDER BY yan tümcesinde ASC veya DESC değiştiricisi sağlanmazsa, sonuçlar ifadeye göre artan düzende sıralanır.

*ASC* : Artan Sıralama
*DESC* : Azalan Sıralama

#### Örnek:
Evler tablosunda banyo sayısı 1'den büyük veya yatak odası sayısı 2 ya da 2'den büyük evlerin tüm özelliklerini azalan sıra ile listele.

SELECT * FROM evler

WHERE banyo>1 OR yatakOdasi>=2

ORDER BY ev_id DESC;

#### Örnek:
Kullanıcı bilgileri tablosundan kilosu 75'den büyük boyu 180'den küçük kişilerin kilo boyu ve yaşını artan sıra ile listele.

SELECT kilo, boy, yas FROM kullanici_Bilgileri

WHERE kilo>75 AND boy<180

ORDER BY yas ASC;

**Not:** ORDER BY yan tümcesini kullanarak sonuç kümenizi sıralarken, tek bir SELECT deyiminde ASC ve DESC özniteliklerini kullanabilirsiniz .

#### Örnek: 
Kullanici_kilosu göre sıralanan tüm kayıtları azalan düzende, ikinci sırayı kullanici_boyu göre artan düzende listelenen bir sorgu yazınız.

SELECT kullanici_kilosu, kullanici_boyu
FROM kullanici_Bilgileri

WHERE kullanici_Adi = 'Ahmet'

ORDER BY kullanici_kilosu DESC, kullanici_boyu ASC;

#### Örnek:
Yaşadikları şehir istanbul yada ankara olan maası 5 bin tl den fazla olan kullaniciların bilgilerini listele.

SELECT *
FROM kullaniclar

WHERE (yasadiklari_Sehir = 'İstanbul' OR yasadiklari_Sehir = 'Ankara')

AND (maas > 5000);


#### DISTINCT
Sonuç kümesinden yinelenenleri(tekrar edenleri) kaldırmak için kullanılır. DISTINCT yan tümcesi yalnızca SELECT deyimleriyle kullanılabilir. Oracle'da DISTINCT yan tümcesi NULL değerleri yok saymaz. Bu nedenle, SQL ifadenizde DISTINCT yan tümcesini kullanırken, sonuç kümeniz ayrı bir değer olarak NULL'u içerecektir.

*Örnek:*
Yaşı 18 den büyük olan kullanıcılarda yinelenenleri sorgula

SELECT DISTINCT ad,boy FROM kullanici_Bilgileri

WHERE yas > 18

ORDER BY kilo;

#### IN KULLANIMI
Bir SELECT, INSERT, UPDATE veya DELETE deyiminde birden çok VEYA koşulu kullanma ihtiyacını azaltmaya yardımcı olmak için kullanılır.

*Örnek:* Tedarikçi isimleri IBM, HP veya Microsoft olanları listele.

SELECT *
FROM tedarikciler

WHERE tedarikciler 

IN ('IBM', 'HP', 'Microsoft');

*Örnek:* ID'si 10, 11, 13 veya 15 olan kulllanıcılara listele.

SELECT *
FROM kullanici

WHERE kullanici_id IN (10, 11, 13, 15);

#### NOT IN KULLANIMI
Not aslında değili ifadesiyle eş değerdir.

*Örnek:*Tedarikçi isimleri IBM, HP veya Microsoft olmayanları listele.

SELECT *
FROM tedarikciler

WHERE tedarikciler NOT IN ('IBM', 'HP', 'Microsoft');

#### BETWEEN KULLANIMI : 
İki değer arasını belirtmek için kullanılır.

*Örnek:*

İletişim tablosundan elmas soyadını taşıyan ve iletişim id sı 1000 ile 2000 arasında olan kullanıcıların bilgilerini listele. 

SELECT * FROM iletişim

WHERE soyadi='elmas' 

AND iletisim_id BETWEEN 1000 AND 2000

#### NOT BETWEEN KULLANIMI

*Örnek:*
kullanıcı id si 4000 ile 4100 arasında olmayan kullanıcıların bilgilerini listele

SELECT *
FROM kullanicilar
WHERE kullanici_id NOT BETWEEN 4000 AND 4100;


#### IS NULL KULLANIMI
NULL boş bir değer olup olmadığını test etmek için kullanılır. Oracle IS NULL koşulunu bir SQL deyiminde veya bir PLSQL kodu bloğunda kullanabilirsiniz.

İfade bir NULL değerse, koşul DOĞRU olarak değerlendirilir.

İfade bir NULL değer değilse, koşul YANLIŞ olarak değerlendirilir.

*Örnek:*

SELECT *
FROM tedarikciler

WHERE tedarikciler IS NULL;



#### IS NOT NULL KULLANIMI
NOT NULL değerini test etmek için kullanılır. Oracle IS NOT NULL koşulunu bir SQL deyiminde veya bir PLSQL kodu bloğunda kullanabilirsiniz.

*Örnek:*

SELECT *
FROM tedarikciler

WHERE tedarikciler IS NOT NULL;

#### LIKE KULLANIMI
LIKE koşulu , bir SELECT , INSERT , UPDATE veya DELETE ifadesinin WHERE yan tümcesinde joker karakterlerin kullanılmasına izin verir . Bu, desen eşleştirmesi yapmanızı sağlar.

**SÖZ DİZİMİ:** expression LIKE pattern [ ESCAPE 'escape_character' ]

expression: Sütun veya alan gibi bir karakter ifadesi.

pattern: Desen eşleştirmeyi içeren bir karakter ifadesi. Aralarından seçim yapabileceğiniz desenler şunlardır:



% : Herhangi bir uzunluktaki herhangi bir dizeyi eşleştirmenize izin verir. (sıfır uzunluk dahil)

_ : Tek bir karakter üzerinde eşleştirmenizi sağlar.

escape character: İsteğe bağlı. % veya _ gibi bir joker karakterin değişmez örneklerini test etmenize olanak tanır.

#### % joker karakter kullanma 

*Örnek:*

Soy adında Mu ile başlayan tüm müşterileri listele

SELECT soyad
FROM müsteriler

WHERE soyadı LIKE 'Mu%';

*Örnek:*
Soy adının içinde mu olan tüm müşterileri listele

SELECT soyad
FROM müsteriler

WHERE soyadı LIKE '%mu%';

#### _ joker karakter kullanma
*Örnek:* 

İlk iki karakterin 'mu' ve son iki karakterin 'lu' olduğu , soyadı 5 karakter uzunluğunda olan tüm kullanıcıları döndürür. Örneğin, soyadı 'mutlu', 'muslu', 'muzlu', 'mudlu' vb. olabilir

SELECT soyad
FROM kullanici

WHERE soyad LIKE 'mu_lu';

*Örnek:* 

Bir telefon numarası aradığınızı fark edebilirsiniz, ancak 11 haneden yalnızca 10 tanesine sahipsiniz. Potansiyel olarak 10 kaydı geri alacaktır. (burada eksik değer 0 ile 9 arasında herhangi bir şeye eşit olabilir). 

SELECT *
FROM kullanıcı

WHERE telefon_numarasi LIKE '053_4441444';

#### NOT LIKE Kullanma

*Örnek:*

NOT Operatörü ile % joker karakterini kullanalım. Adı 'S' ile başlamayan tedarikçileri bulmak için Oracle LIKE koşulunu da kullanabilirsiniz.

SELECT ad
FROM kullanici
WHERE ad NOT LIKE 'S%';

#### Kaçış Karakterlerini Kullanma
Oracle LIKE koşulunda bir % veya _ karakteri aramak istediğinizi varsayalım. Bunu bir Escape karakteri kullanarak yapabilirsiniz.

*Örnek:* 

! bir kaçış karakteridir. Bu bildirim, adı Water% olan tüm kullanıcıları döndürür.

SELECT *
FROM kullanici

WHERE ad LIKE 'Mutlu!%' ESCAPE '!';

*Örnek:* 

Bu LIKE koşulu örneği, adı H ile başlayan ve % ile biten tüm kullanıcıları döndürür. Örneğin 'Hayriye%'

SELECT *
FROM kullanici

WHERE ad LIKE 'H%!%' ESCAPE '!';

*Örnek:*

Bu LIKE koşulu örneği, adı H ile başlayan ve _ ile biten tüm kullanıcıları döndürür. Örneğin 'Hayriye_'

SELECT *
FROM kullanici

WHERE ad LIKE 'H%!_' ESCAPE '!';

*Örnek:*

"Test" kelimesini içeren tüm kayıtlar için bir serbest metin alanına karşı sorgulama yapmaya çalışıyorum. Sorun şu şekilde girilebilmesidir: TEST, Test veya test.

SELECT *
FROM kullanici

WHERE UPPER(ad) LIKE ('TEST%'); 

ifadesiyle TEST, Test veya test içeren tüm kayıtlara bakabilirim.

Aşağıdaki tabloda Adından H harfi bulunan çalışanların sorgulaması yer almaktadır.






#### REGEXP KULLANIMI

Bir SELECT, INSERT, UPDATE veya DELETE ifadesinin WHERE yan tümcesinde düzenli ifade eşleştirmesi gerçekleştirmenize olanak tanır.

Oracle/PLSQL'deki REGEXP_LIKE koşulunun sözdizimi şöyledir:

REGEXP_LIKE ( expression, pattern [, match_parameter ] )

expression : Sütun veya alan gibi bir karakter ifadesi. VARCHAR2, CHAR, NVARCHAR2, NCHAR, CLOB veya NCLOB veri tipi olabilir.

pattern: Normal ifade eşleştirme bilgileri. Aşağıdakilerin bir kombinasyonu olabilir:




match parameter : İsteğe bağlı. REGEXP_LIKE koşulu için eşleşen davranışı değiştirmenize olanak tanır. Aşağıdakilerin bir kombinasyonu olabilir:



*Örnek:*
Kullanıcı adı Yasan, Yasen yada Yasin olan kullanıcıları listele.

SELECT ad
FROM kullanicilar
WHERE REGEXP_LIKE (ad, 'Yas(a|e|i)n');

*Örnek:*
soyadı 'A' ile başlayan tüm kişileri lsitele.

SELECT soyad
FROM kullanicilar
WHERE REGEXP_LIKE (soyad, '^A(*)');

*Örnek:*
soyadı 'n' ile biten tüm kişileri listele.

SELECT soyad
FROM kullanicilar
WHERE REGEXP_LIKE (soyad, '(*)n$');

#### ALİASLAR
Sütunlar veya tablolar için geçici bir ad oluşturmak için kullanılabilir.

* **COLUMN ALIAS**ları, sonuç kümenizdeki sütun başlıklarının daha kolay okunmasını sağlamak için kullanılır.

* **TABLO ALIAS**ları, okumayı kolaylaştırmak için veya kendi kendine birleştirme gerçekleştirirken SQL'inizi kısaltmak için kullanılır.

Oracle/PLSQL'de ALIAS A COLUMN sözdizimi şöyledir:

column_name AS alias_name

Oracle/PLSQL'de ALIAS A TABLE sözdizimi şöyledir:

table_name alias_name

#### ALIAS bir Sütun

*Örnek:* ad ve soyadı birleştirirek isim adında bir sütün yapalım.

SELECT baglantı_id, ad || soyad AS "isim"
FROM baglantılar
WHERE soyad = 'yilmaz';

#### ALIAS bir Tablo

SELECT a.kullanici_id, a.kullanici_isim, b.calistigiAlan_isim
FROM kullanici_Bilgileri a
INNER JOIN calisma_Bilgileri b
ON a.calistigiAlan_id = b.calistigiAlan_id
ORDER BY a.kullanici_id ASC, b.calistigiAlan_isim ASC ;


**INNER JOİN** : İki veya daha fazla tablodan ortak olan iki alandaki değerleri kontrol ederek tabloları birleştirme işlemidir.

*Örnek* :

SELECT evler.evler_id, saticilar.soyad, saticilar.ad
FROM saticilar

INNER JOIN evler

ON saticilar.saticilar_id = evler.saticilar_id

ORDER BY evler_id;
